```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

## Distributon Plot
- visualization of distribution of a data set
- use one column

```python
tips = sns.load_dataset("tips")
```

```python
tips.info()
```

```python
sns.displot(tips["total_bill"], kde = True, bins = 50)
```

## Joint Plot
- allows you to combine two different distribution plots
- using to compare in a scatter
- hex as a kind: tells you where more of your points are
- reg as a kind: Liner fit 
- kde as a kind: concentrated point

```python
sns.jointplot(x = "total_bill" , y = "tip" , data = tips)
```

### Pair Plot
- plots relations across an entire data frame, performs a joint for all data colums combinations

```python
sns.pairplot(tips, hue="smoker", palette="twilight")
```

## Categorical Plots

#### bar Plot
- default is mean but we can change to max
- x is a categorical column

```python
sns.barplot(x = "sex", y = "total_bill", data = tips, estimator= "max")
```

### Count Plot
- only a x value-> JUst coun the occurrances

```python
sns.countplot(x = "sex", hue="smoker", data = tips, palette="magma")
sns.despine()
```

## Box Plots

```python
sns.boxplot(x = "day", y = "total_bill", data = tips, hue="smoker")
```

## Matrix plots

```python
flights = sns.load_dataset("flights")
```

```python
flights
```

## Heatmap

```python
sns.heatmap(tips.corr(), annot=True, lw=2)
```

```python
flights.head()
```

## Flights an matrix

```python
fp = flights.pivot_table(index="month", columns="year", values = "passengers")
```

```python
plt.figure(figsize=(16,6))
sns.heatmap(fp)
```

## Cluster Map

```python
sns.clustermap(fp, cmap = "coolwarm", standard_scale=1)
```
