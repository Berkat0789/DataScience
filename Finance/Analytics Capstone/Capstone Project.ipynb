```python
import pandas as pd
import numpy as np 
import yfinance as yf 
import seaborn as sns
```

```python
tickers = ["TSLA", "GM", "F"]
```

```python
df = pd.DataFrame()
for x in tickers:
    df[x] = yf.download(x, start="2019-1-1", progress=False)["Adj Close"]
```

## Daily Returns

```python
for x in tickers:
    df[f"{x} daily return"] = df[x].pct_change()
```

```python
df
```

```python
df["TSLA daily return"].hist(bins=100)
```

## Cumulative Returns

```python
for x in tickers:
    df[f"{x} cumulative Return"] = (1 + df[f"{x} daily return"]).cumprod()
```

```python
df
```

```python
df[["TSLA cumulative Return", "GM cumulative Return", "F cumulative Return"]].plot()
```
