## Creating Model
- Checking for unbalanced target
- if so we will need to resample: see below

```python
sns.countplot(x = df["booking_status"])
```

```python
target = "booking_status"
```

```python
X = df.drop(target, axis=1).values 
y = df[target].values
```

## SMOTE
- Because the bookstatus values are unbalanced we need to resample to fit the target

```python
X, y = SMOTE().fit_resample(X,y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scalar = MinMaxScaler()
```

```python
X_train = scalar.fit_transform(X_train)
```

```python
X_test = scalar.transform(X_test)
```

```python
X_train.shape
```

```python
model = Sequential()
model.add(Dense(units=120, activation="relu"))
model.add(Dense(units=120, activation="relu"))
model.add(Dense(units=120, activation="relu"))
model.add(Dense(units=120, activation="relu"))
model.add(Dense(units=100, activation="relu"))
model.add(Dense(units=50, activation="relu"))
model.add(Dense(units=40, activation="relu"))

model.add(Dense(units=1, activation="sigmoid"))

model.compile(loss = "binary_crossentropy", optimizer = "adam", metrics = ["accuracy"])
```

```python
stop  = EarlyStopping(patience=60)
```

```python
model.fit(X_train, y_train, epochs = 200, validation_data = (X_test, y_test), callbacks = [stop])
```

```python
pd.DataFrame(model.history.history)[["loss", "val_loss"]].plot()
```

```python
pd.DataFrame(model.history.history)[["accuracy", "val_accuracy"]].plot()
```

## Deep Network Result
- Produces 90% at first test will run again on kaggle

```python
predictions = model.predict_classes(X_test)
```

```python
print(classification_report(predictions, y_test))
```

```python
print(confusion_matrix(predictions, y_test))
```

```python
print(explained_variance_score(predictions, y_test))
```
