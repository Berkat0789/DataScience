```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.image import imread
import os
import seaborn as sns
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPool2D, Flatten, Reshape, Dropout
from tensorflow.keras.optimizers import Adam, SGD
from tensorflow.keras.losses import CategoricalCrossentropy
```

```python
mainPath = "gems/"
```

```python
os.listdir(mainPath)
```

```python
trainPath = mainPath + "train"
```

```python
testPath = mainPath + "test"
```

```python
os.listdir(testPath + "/Coral/")
```

```python
sampleImagePath = testPath + "/Coral/" + "coral_18.jpg"
```

```python
sampleImage = imread(sampleImagePath)
```

```python
plt.imshow(sampleImage)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
imageGen = ImageDataGenerator(rotation_range=45,
                              height_shift_range=0.2,
                              width_shift_range=0.2,
                              zoom_range=0.65,
                              vertical_flip=True,
                              horizontal_flip=True,
                              shear_range=0.2
                             )
```

```python
plt.imshow(imageGen.random_transform(sampleImage))
```

```python
testGenerator = imageGen.flow_from_directory(testPath,
                                             target_size=(150,150),
                                             batch_size=363,
                                             class_mode="categorical",
                                             shuffle=False
                                            )
```

```python
trainGenerator = imageGen.flow_from_directory(trainPath,
                                             target_size=(150,150),
                                             batch_size=2856,
                                             class_mode="categorical",
                                             shuffle=True
                                            )
```

```python
X_test, y_test = testGenerator.next()
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
X_train, y_train = trainGenerator.next()
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_train = X_train/255
```

```python
X_test = X_test/255
```

```python
model = Sequential()
model.add(Conv2D(filters=32, kernel_size=(2,2), strides=(1,1), input_shape = (150,150,3), activation="relu"))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Conv2D(filters=32, kernel_size=(2,2), strides=(1,1), input_shape = (150,150,3), activation="relu"))

model.add(Flatten())

model.add(Dense(450, activation="relu"))
model.add(Dense(450, activation="relu"))
model.add(Dense(150, activation="relu"))
model.add(Dense(87, activation="softmax"))

model.compile(loss = CategoricalCrossentropy(), optimizer = Adam(), metrics = ["accuracy"])
```

```python
model.fit(X_train, y_train, epochs = 10, validation_data = (X_test, y_test))
```
