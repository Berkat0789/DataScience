## Average Size of the images 
- used to set a relative inputShape

```python
dim1 = []
dim2 = []

for imageName in os.listdir(trainPath + "/normal"):
    img = imread(trainPath + "/normal/" + imageName)
    d1,d2 = img.shape[:2]
    dim1.append(d1)
    dim2.append(d2)
```

```python
np.mean(dim1)
```

```python
np.mean(dim2)
```

```python
inputShape = (200,200,3)
```

## Generator

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
generator = ImageDataGenerator(rotation_range=90,
                               vertical_flip=True,
                               horizontal_flip=True,
                               height_shift_range=0.2,
                               width_shift_range=0.2,
                               shear_range=0.15,
                               zoom_range=0.60
                              )
```

```python
plt.imshow(generator.random_transform(sampleImage))
```

```python
trainGenerator = generator.flow_from_directory(trainPath,
                                               target_size= inputShape[:2],
                                               class_mode="binary",
                                               color_mode="rgb",
                                               shuffle=True,
                                               batch_size=3800
                                              )
```

```python
testGenerator = generator.flow_from_directory(testPath,
                                               target_size= inputShape[:2],
                                               class_mode="binary",
                                               color_mode="rgb",
                                               shuffle=True,
                                              batch_size=400
                                              )
```

```python
X_train, y_train = trainGenerator.next()
```

```python
X_test, y_test = testGenerator.next()
```

## Scaling

```python
X_train = X_train/255
```

```python
X_test = X_test/255
```

```python
model = Sequential()
model.add(Conv2D(filters=31, kernel_size=(2,2), strides=(1,1), padding="same", activation="relu"))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Conv2D(filters=31, kernel_size=(2,2), strides=(1,1), padding="same", activation="relu"))

model.add(Flatten())

model.add(Dense(units=250, activation="relu"))
model.add(Dense(units=150, activation="relu"))
model.add(Dense(units=150, activation="relu"))
model.add(Dense(units=1, activation="sigmoid"))

model.compile(loss = BinaryCrossentropy(), optimizer = Adam(), metrics = ["accuracy"])
```

```python
model.fit(X_train, y_train, epochs = 10, validation_data = (X_test, y_test))
```
