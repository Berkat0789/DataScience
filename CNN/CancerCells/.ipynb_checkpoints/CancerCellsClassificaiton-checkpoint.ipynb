```python
inputShape = (120,120,3)
```

## Generator

```python
imageGenerator  = ImageDataGenerator(rotation_range=40,
                                     height_shift_range=0.3,
                                     width_shift_range=0.3,
                                     horizontal_flip=0.8,
                                     vertical_flip= 0.8,
                                     shear_range= 0.4,
                                     zoom_range=0.6,
                                     fill_mode="nearest",
                                     rescale= 1/255
                                    )
```

```python
plt.imshow(imageGenerator.random_transform(sampleImage))
```

## Train Generator

```python
trainGenerator = imageGenerator.flow_from_directory(trainPath,
                                                    target_size=inputShape[:2],
                                                    class_mode="binary",
                                                    batch_size=12,
                                                    shuffle=True
                                                   )
```
