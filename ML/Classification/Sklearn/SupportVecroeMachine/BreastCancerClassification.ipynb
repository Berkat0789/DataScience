```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
bc = load_breast_cancer()
```

```python
df = pd.DataFrame(bc["data"], columns=bc["feature_names"])
```

```python
df["target"] = bc["target"]
```

```python
df
```

```python
plt.figure(figsize=(16,6))
sns.heatmap(df.corr(), annot=True, linewidths=3, linecolor="black")
```

```python
df.corr()
```

```python
sns.jointplot(x = "mean concave points" ,y = "mean radius", data=df, hue="target")
```

```python
X = df.drop("target", axis=1)
y = df["target"]
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, explained_variance_score
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
print(explained_variance_score(y_test,predictions))
```

### Modify Parameters
- C or gamma values 

```python
from sklearn.model_selection import GridSearchCV
```

```python
paramGrid = {"C": [0.1,1,10,100,1000], "gamma": [1,0.1,0.001,0.0001]}
```

```python
grid = GridSearchCV(SVC(), paramGrid, verbose=3)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
gridPredict = grid.predict(X_test)
```

```python
print(classification_report(y_test,gridPredict))
```
